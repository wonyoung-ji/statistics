## 4.1.속성수준별 효용가치 계산
- 속성수준을 표현하는 더미변수 생성
- 응답자별 효용가치를 구하기 위해 회귀분석 수행
- 이를 속성수준별로 합산

```
# 속성수준 더미변수생성 및 독립변수 객체 생성

import statsmodels.api as sm
from sympy import *

con_df = pd.get_dummies(df, columns=['데이터양','데이터속도','통화량'], drop_first=False)
con_df['intercept'] = 1.0
X = con_df[['intercept','데이터양_1GB','데이터양_2GB','데이터속도_중','데이터속도_상','통화량_150분','통화량_180분']]
X.head()
```

- get_dummies(): 더미변수 만들기. drop_first=False로 지정하여 모든 속성수준에 더미변수 만듦

```
# 개별 응답자의 속성수준 효용가치 계산
res = []
intercept = []
for i in range(50):
    Y = con_df[con_df.columns[2:52][i]] # 응답자1~50까지 반복
    LR1 = sm.OLS(Y, X).fit()

# 연립방정식을 이용한 응답자의 속성수준별 효용가치 계산 
  # 데이터양
    x, y, z= symbols('a_1 a_2 a_3')
    a1=solve( [ Eq(x-z,LR1.params[1]), Eq(y-z,LR1.params[2]), Eq(x+y+z,0)], [x,y,z] )
  # 데이터속도
    x, y, z = symbols('b_1 b_2 b_3')
    b1=solve( [ Eq(x-z,LR1.params[3]), Eq(y-z,LR1.params[4]), Eq(x+y+z,0)], [x,y,z] )
  # 통화량
    x, y, z = symbols('c_1 c_2 c_3')
    c1=solve( [ Eq(x-z,LR1.params[5]), Eq(y-z,LR1.params[6]), Eq(x+y+z,0)], [x,y,z] )

# 속성수준 효용가치 병합
    a_1=list(a1.values())
    b_1=list(b1.values())
    c_1=list(c1.values())
    a_1.extend(b_1)
    a_1.extend(c_1)
    res.append(a_1)
    d_1=LR1.params[0]
    intercept.append(d_1)
```

- OLS(): 회귀분석 수행 함수
- res = 개별응답자들의 속성수준별 효용가치가 저장될 리스트 
- intercept = 절편이 저장될 리스트

- 회귀분석 결과로 출력된 회귀계수를 이용하여 연립방정식을 세우고, 이를 통해 속성수준별 효용가치를 추정
- symbol(): 방정식에 사용될 변수 선언. 변수간 띄어쓰기로 구분
- Eq(좌변, 우변): 방정식 기재
- solve(식, 변수): 연립방정식의 해 출력


- 딕셔너리 형태로 되어 있는 a1, b1, c1들의 values를 출력해서 list 형태로 만듦
- 하나의 리스트로 합치기 위해 extend()사용
- 최종적으로 만들어진 a_1을 res에 저장

## 4.4.최대선호 모형 기반 시장점유율 예측
- 응답자별 속성 수준의 효용가치 값을 이용하여 시장점유율 예측 가능
- 최대선호 모형, 선택확률 모형

```
# 9가지 대안에 대한 개별응답자의 최대선호 계산
pre1 = []
for i in range(0,50):
  market_sum1 = np.where(cust_Utility.loc[i].values == cust_Utility.loc[i].max(),1,0)
  pre1.append(market_sum1)
MAX = pd.DataFrame(pre1)
MAX.columns = cust_Utility.columns
MAX.head()
```

```
# 최대선호 모형을 이용한 시장점유율 예측
pre2 = []
for i in range(0,9):
  MAX_sum1 = MAX.iloc[:,i].sum()
  pre2.append(MAX_sum1)
prefer = pd.DataFrame(pre2).T
prefer.columns = cust_Utility.columns
prefer/50*100
```

- for문: 대안별 합산
- 시장점유율을 계산하기 위해 prefer 데이터를 전체 응답자 수인 50으로 나누어 비율을 구함
- 최적 대안 우선순위에서 __대안 8이 2순위 였음에도 불구하고, 시장점유율은 0__이 출력
  - 최대선호라는 방식은 __가장 높은 효용가치를 주는 대안만 선택하고 나머지는 버려진다는 전제__를 가지고 있기 때문
