# __신뢰성 분석__

# 1.신뢰성과 타당성
- 신뢰성(reliability): 측정도구의 안정성, 일관성, 예측가능성, 정확성, 의존가능성
- 타당성(validity): 측정도구의 정확성 또는 적절성
  - 타당성이 만족되기 위해선 신뢰성이 만족되어야 함
  - 신뢰성이 높다고 해서 타당성이 높은 것이 아님

## 1.1.측정오차
 - 측정오차 = __실제값 - 측정값__ 
  + 실제값 = 원래 나의 성적
  + 측정값 = 컨디션에 따른 시험 당일 성적

- 측정오차 = 체계적 오차 + 비체계적 오차
  + __체계적 오차__: 일정한 패턴을 나타내는 오차. 타당성 결정. 
        예) 사격수의 총기 자체의 품질.
            이에 따라 과녁 가운데는 빗나가지만, 매번 같은 곳을 맞힘.
  + __비체계적 오차__: 일정한 패턴이 없는 오차. 신뢰성 결정.
          예) 사격수의 실력 or 사격환경의 품질.
              아무리 좋은 총기를 지녔다고 하여도 실력이 없거나, 거센 바람이 불면 중앙을 벗어나게 됨.
              이에 따라 과녁 가운데를 맞춰도 탄착점은 흩어져 나타남.



image.png

- 1번: 신뢰성↑ + 타당성↓. 실력있는 명사수와 딱총. 중앙에선 벗어난, 항상 같은 곳에 맞춤
- 2번: 신뢰성↑ + 타당성↑. 실력있는 명사수와 활. 중앙에 위치한, 항상 같은 곳에 맞춤
- 3번: 신뢰성↓ + 타당성↓. 실력없는 꼬마와 딱총. 중앙에서 벗어난, 항상 다른 곳에 맞춤
- 4번: 신뢰성↓ + 타당성↑. 실력없는 꼬마와 활. 중앙에 위치한, 항상 다른 곳에 맞춤

## 1.2.신뢰성
- __신뢰성__: 측정대상을 __반복적으로 측정__했을 때 동일한 결과가 나오는 정도
- 비체계적 오차의 통제와 관리가 중요
- __반복측정, 대체측정, 반분측정, 내적일관성__ 등

- __반복측정__
  + 측정방법이 하나밖에 없을 경우
  + __시간적인 간격을 두고 2회__에 걸쳐 측정 후 상관관계 분석
  + 단점
        1. 시간의 간격이 신뢰성 평가에 영향을 미칠 수 있음
        2. 추가적인 조사 및 측정으로 인해 시간과 비용이 큼
        3. 피실험자들의 숙지가 두 번째 측정에 영향을 줄 수 있음 (주시험효과)



- __대체측정__(=동형검사법)
  +  대등한 두 가지의 측정 도구가 있을 경우
  + __동시에 동일한 측정 대상__을 측정 후 상관관계 분석
  + 장점
        반복측정의 단점을 보완
  + 단점
        시간과 비용이 많이 요구됨



- __반분측정__
  + 다수의 측정 항목이 존재할 경우
  + __측정항목__을 두 그룹으로 나누고, __측정 대상__ 역시 __두 그룹__으로 나눈 후, 상관관계 분석
  + 장점:
        보다 정확한 신뢰성
  + 단점:
        1.측정항목을 두 그룹으로 분리할 만큼, 충분한 개수 필요
        2.여러 항목의 동일한 개념을 측정하는 측정도구의 동질성 확보 필요


- __내적일관성__
  + 하나의 개념을 측정하기 위해 여러 개의 하위 측정항목들이 존재할 경우
  + 크론바흐 알파(Cronbach's Alpha)를 바탕으로 측정 항목간의 상관관계 분석
  + 신뢰도를 저하시키는 하위 측정항목을 찾아 제거하여 전체적인 신뢰성 높임
  + 가장 많이 활용되는 신뢰성 평가기법
  + 고려할 사항
        1. 구성개념(측정하고자 하는 개념과 하위 측정항목간의 관계)을 명확히 이해
        2. 신뢰성이 높다고 평가된 기존의 측정항목을 사용
        3. 측정항목의 수는 최소 3개 이상
        4. 내적일관성 분석 후 전체적인 신뢰도를 저해하는 항목 제거
        5. 측정자 or 실험자의 실험방식 및 측정환경 일관되도록 유지



## 1.3.타당성
- __타당성__: 구성개념(측정하고자 하는 대상, construct)를 __측정항목이 제대로__ 혹은 __올바르게 측정__하고 있는지의 정도
- 신뢰성과 달리, 타당성을 정확히 평가할 수 있는 방법 존재하지 않음
- __내용타당성, 기준타당성, 구성타당성__

image.png

- __내용타당성__
  + 측정도구가 측정하고자 하는 __개념의 내용이나 속성을 얼마나 정확히 측정__하고 있느냐의 수준을 의미
  + 측정도구가 포함하고 있는 내용적인 측면 판단
  + 통계적으로 규명 불가능

- __기준타당성__
  + 특정 측정도구의 측정 결과값을 기준으로 삼아 __실제로 평가하고자 하는 개념이나 상태를 올바르게 나타내고 있는지__의 정도를 의미
  + __먼저 측정된 측정결과 값(기준값)과 추후 측정된 측정결과 값의 상관관계__를 바탕으로 평가
  + 통계적으로 평가 가능
  + __예측적 타당성(predictive validity)__
    + 특정 실험의 결과 값으로 예측된 기준이 추후 행위 또는 결과 값과 얼마나 연괸되어 있느냐를 평가
    + 예: 대학수능시험 결과와 실제 대학의 학점간의 상관관계
  + __동시적 타당성(concurrent validity)__
    + 특정 실험의 결과 값으로 예측된 기준이 또 다른 현재 행위 또는 상황의 결과값과 얼마나 연관되어 있느냐를 평가
    + 예: 토익 필기시험과 영어대화능력 평가결과와의 상관관계

- __구성타당성__
  + 측정하고자 하는 __개념의 하위 요소들이 상위 개념을 적절히 측정하고 있는지__를 나타내고 있는 정도
  + 이론적 연구에서 가장 중요한 타당성
  + __집중타당성(convergent validity)__
    + 동일한 개념을 측정하는 다수의 측정도구 사이의 상관관계 정도
    + 싱관관계가 높을 수록 집중타당성이 높음
    + 예: 서비스만족도를 평가하기 위한 세부 측정항목들 사이의 상관관계
  + __판별타당성(discriminant validity)__
    + 상이한 개념을 측정하는 서로 다른 측정도구 사이의 상관관계 정도
    + 상관관계가 낮을수록 판변타당성이 높음
    + 예: 상품만족도를 평가하기 위한 세부 측정항목들과 서비스만족도를 평가하기 위한 세부 측정항목들 사이의 낮은 상관관계
  + __이해타당성(nomological validity)__
    + 구성개념들간의 이론적인 혹은 논리적인 관계가 실제 측정 결과에서도 나타나고 있는지를 판단
    + 구성개념들간의 관계를 가설 및 검정하여 판단
    + 예: 직원만족도가 고객만족도에 영향을 미친다는 이론적 관계는 실제 측정된 직원만족도와 고객만족도 값 간의 가설검정을 통해 규명됨
  + __상관관계표__를 활용 or __요인분석__ 사용하여 __집중타당성__과 __판별타당성 판단__
  + 이론적인 관계를 규명하기 위한 __별도의 가설검증__을 통해 __이해타당성 판단__

# 2.내적일관성 분석

## 2.1.개념 및 주요 이론
- 내적일관성 분석: 신뢰성을 측정하는 방법 중 하나
- 하나의 개념을 측정하기 위해 여러 개의 __하위 측정항목들__이 존재할 경우 이들 측정 값 __사이의 상관관계를 분석__하여 신뢰성을 평가하는 방법
- 크론바흐 알파(Cronbach's Alpha)를 사용하여 계산


image.png

- __α 0.6 이상__ = 신뢰성 존재
- __α 0.7 이상__ = 신뢰성 높음
- α가 낮을 경우 __측정 항목들간 상관관계가 낮은 측정항목을 제거__하여 α를 향상시킴
- 하나의 개념에 대한 __신뢰성 및 타당성을 평가하기 위한 최소의 항목 수는 3개__이므로 모든 개념의 측정 항목 수는 3개 이상으로 유지해야 함

## 2.2.분석 및 해석
- 쇼핑몰은 고객 직원에 대한 만족도 평가하기 위해 친절성, 신속성, 책임성, 정확성, 전문성을 측정하는 측정항목 개발
- 5개 항목으로 구성된 직원 만족도 평가가 신뢰성 있는 조사인지 확인하기 위해 내적일관성 분석 수행

```
import pandas as pd
import pingouin as pg
df = pd.read_csv('/content/gdrive/My Drive/통계/Ashopping.csv',sep=',',encoding='CP949')
df.head()
```

```
X = df[['친절성','신속성','책임성','정확성','전문성']]
X.head()
```

```
# 크론바흐 알파 계수 출력
pg.cronbach_alpha(data=X)
```

- 크론바흐 알파 계수 = 0.811
- [0.793, 0.83]은 신뢰수준 95%의 신뢰구간을 의미

```
# 각 변수를 제외한 크론바흐 알파 계수 출력
X1 = df[['신속성','책임성','정확성','전문성']]
X2 = df[['친절성','책임성','정확성','전문성']]
X3 = df[['친절성','신속성','정확성','전문성']]
X4 = df[['친절성','신속성','책임성','전문성']]
X5 = df[['친절성','신속성','책임성','정확성']]

print(pg.cronbach_alpha(data=X1))
print(pg.cronbach_alpha(data=X2))
print(pg.cronbach_alpha(data=X3))
print(pg.cronbach_alpha(data=X4))
print(pg.cronbach_alpha(data=X5))
```

- 모두 0.811 미만으로 출력
- 어떠한 변수를 제거하여도 크론바흐 계수는 높아지지 않음
- 즉, 측정항목의 변화 없이 5개 모두를 사용하는 것이 적절
- 현재의 측정항목들로 구성된 직원 만족도 평가는 신뢰성이 있는 측정도구 구성되어 있음
