# __요인분석__

# 1.요인분석 기초
- 요인분석
  + 다수의 변수들간의 상관관계를 바탕으로, __자료들을 보다 이해하기 쉬운 형태로 축소 또는 요약__할 때 사용하는 방법
  + 관련 변수들이 묶여 요인을 이룸
  + __요인은 상호독립적__ 특성을 가지므로 구조적 특성을 명확히 파악 가능
  + 관련성이 없거나 중요도가 낮은 변수들을 선별하여 제거 가능
  + 동일한 요인으로 묶였는지 확인하여 __타당성 평가__ 가능
  + 요인분석 후 얻은 요인들을 __회귀분석, 판별분석 등으로 활용__ 가능


## 1.1.요인추출방방법
- 변수들 내부에 존재하는 구조를 발견하고자 할 때 __상관관계를 바탕으로 변수들을 대표할 수 있는 요인 구조를 추출__
- __주성분분석(Principle Component Analysis)__
  + 예측을 목적으로 수집된 자료들을 가능한 최소의 요인으로 축소할 때
  + 공통변량과 고유변량 모두 고려. __총변량(Total Variance)__에 기초하여 요인 추출
  + 방법
        1. n개의 변수들이 가지는 총분산 n개의 주성분으로 추출
        2. 먼저 추출되는 주성분요인일수록, 총분산을 많이 설명하도록 순차적으로 추출
        3. 차원의 축소가 목적이기 때문에, 모든 주성분요인을 사용하는 것이 아닌, 가장 많은 양의 분산을 설명하는 주성분요인만 선별
- __공통요인분석(Common Factor Analysis)__
  + 변수들 사이에 존재하는 차원이나 요인들을 발견하기 위한 분석목적일 때
  + 고유변량을 제외하고 __공통변량(Common Variance)__에만 기초하여 요인 추출
  + 마찬가지로, 공통분산을 설명하는 모든 요인을 추출하지 않음
  + 대신, 많은 양의 공통분산을 설명하는 요인을 우선적으로 추출

## 1.2.요인분석의 주요지표

### 1.2.1.요인적재량(Factor Loading)
+ __각 변수와 각 요인간의 관련성 정도__를 나타내는 값
+ 측정변수 = 종속변수
+ 추출된 각 요인 = 독립변수
+ (표준화된 자료를 사용할 경우, 요인적재량은 단순히 상관관계 값)
+ 요인적재량이 0.3이상이면 유의한 관계
+ __요인적재량이 가장 큰 요인으로 해당 변수가 소속됨__

### 1.2.2.공통성(Communality)
  + __특정 변수에 대해 추출된 요인들이 설명하는 비율__
  + 공통성 = 요인들이 갖는 요인적재량을 제곱한 값들의 합
  + 0.5이상 = 공통성 충분히 가지고 있음 
  + 공통성이 0.5이하인 변수는 제외시키는 것이 적절

image.png

###1.2.3.고유값(Eigenvalue)
- 개별 측정변수의 변량을 1이라고 했을 때, __추출된 각 요인이 갖는 변량의 크기__
- 측정된 변수들의 변량 중, 한 요인에 의해 설명되는 분산의 양
- __특정요인이 측정변수를 몇 개나 설명하는지를 의미__
  + ex) 특정요인의 고유값이 4.5라면, 해당 요인은 4.5개의 개별 측정변수만큼의 변량을 설명하고 있음




image.png

### 1.2.4.요인점수(Factor Score)
- 각각의 사례가 추출된 요인을 반영하고 있는 정도(__사례와 각 요인이 갖는 관계의 정도__)
- 각 변수들의 표준화된 값을 요인점수 계수행렬에 곱해서 구함
- 요인점수 값을 통해 요인수준의 회귀분석, 분산분석 등 수행

## 1.3.요인의 회전
- 요인적재량이 요인 별로 유사하게 나타날 경우 해당변수가 어느 요인에 속하는지 판단하기 어려움
- 요인회전 = __요인의 축을 이동시켜 변수들이 갖는 각 요인들에 대한 요인적재량을 명확하게 나타내는 방법__
- __직각회전__
  + 요인들간의 상호독립성을 유지하도록 회전시킴
  + 다중공선성을 제거할 수 있음
- __사각회전__
  + 요인들이 서로 상관관계가 없다는 가정을 하지 않고 회전
  + 직각을 유지하지 않으므로, 높은 요인적재량은 더 높게, 낮은 요인적재량은 더 낮게 출력

## 1.4.요인분석의 유형
- __탐색적 요인분석__
  + 측정된 변수들간의 관계성에 대한 사전지식이나 이론적 배경 없이 요인을 추출
  + 원칙적으로 몇 개의 요인이 도출될 지 알 수 없음
- __확인적 요인분석__
  + 측정된 변수들간의 관계성에 대한 사전지식이나 이론적 배경이 분명
  + 요인 수를 강제로 설정하고, 요인과 변수간의 구조적 특징이 적합한지 판단

# 2.탐색적 요인분석

## 2.1.개념 및 주요 이론
- __탐색적 요인분석__
  + 변수들 간의 관계에 대한 이론이 아무것도 정의되지 않은 상태
  + 변수들의 이면에 있는 공통요인들간의 관련성을 알아보기 위해 탐색적으로 요인을 파악
  + 분석결과로 도출된 요인들이 몇 개 일지, 세부 측정변수들은 무엇일지 미리 알 수 없음
  + 상향식(Bottom-up)요인분석 방식이라고도 함
- __직교요인 모형__
         1.각 변수들은 (공통)요인들의 선형결합 및 고유요인에 의해 나타남. 따라서, 공통요인들은 모든 변수에 영향을 주지만 각 고유요인은 해당 변수에만 영향을 줌
         2. 공통요인들은 서로 독립('직교'라 불리는 이유)
         3. 고유요인들은 서로 독립
         4. 공통요인과 고유요인은 서로 독립

## 2.2.분석 및 해석
- 쇼핑몰 고객만족도 조사는 상품품질 ~ 안내표지판 설명 총 10개의 항목
- 개별 항목들도 중요한 항목이지만, 효율적인 측정지표 관리를 위해 탐색적 요인분석 실시

```
import pandas as pd
from factor_analyzer import FactorAnalyzer
df = pd.read_csv('/content/gdrive/My Drive/통계/Ashopping.csv',sep=',',encoding='CP949')
df.head()
```

- 요인분석을 위해 factor_analyzer 패키지 설치

```
# 변수 추출
X = df[['상품_품질','상품_다양성','가격_적절성','상품_진열_위치','상품_설명_표시',
        '매장_청결성','공간_편의성','시야_확보성','음향_적절성','안내_표지판_설명']]
X.head()
```

```
# 탐색적요인분석
fa = FactorAnalyzer(method = 'principal',n_factors=10,rotation='varimax').fit(X)
```

- FactorAnalyzer(): 요인분석 지원 함수
  - method = 요인추출방법
  - n_factors = 축약하고자 하는 요인의 수(default=3)
  - rotation = 요인 회전 방식

```
# 결과출력
print('요인적재량')
pd.DataFrame(fa.loadings_, index=X.columns)
```

```
print('공통성')
pd.DataFrame(fa.get_communalities(), index=X.columns)
```

```
print('고유값')
ev,v = fa.get_eigenvalues()
pd.DataFrame(ev)
```

```
print('요인점수')
print(fa.transform(X.dropna()).shape)
fa.transform(X.dropna())
```

- 요인적재량: loadings_ 속성 사용
- 공통성: get_communalities() 함수 사용
- 고유값: get_eigenvalues() 함수 사용
  + 고유값(ev), 공통요인 고유값(v) 두 개 중 고유값만 출력
- 요인점수:transform() 함수 사용


- 탐색적 요인분석 결과
  + 요인적재량을 통해 각각의 변수가 어떤 요인에 속하는지 확인 가능
  + 고유값 기준으로 1보다 큰 값을 갖는 요인은 2개. (총 2개의 요인이 적절)
- 결과 해석
  + 첫번째 요인(매장만족도): 매장청결성, 공간편의성, 시야확보성, 음향적절성, 안내 표지판 설명
  + 두번째 요인(상품만족도): 상품 품질, 상품 다양성, 가격 적절성, 상품 진열 위치, 상품설명 표시
  + __요인점수가 모든 고객에 대해 산출__되었으므로 상품만족도와 매장 만족도를 별개의 변수로 활용하여 다변량 통계분석에 활용가능
