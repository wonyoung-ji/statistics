  - __④ 좌표계 맵핑__

### 2.1.3.차원 수 선정 방법
  - 요인분석과 유사하게, 다양한 변수(속성)을 적은 수의 차원으로 요약하여 분석 대상 개체들의 관계를 표현
  - __크루스칼의 스트레스 값__ 사용
    - 크루스칼의 스트레스 값: 다차원 정보가 저차원으로 요약(축소)됨에따라 발생하는 정보의 손실율
   - image.png
   - 크루스칼의 스트레스값은 0 ~ 1 사이의 값을 가짐. 0에 가까울 수록 적합도 높음
   - __0.1이하일 경우 차원 수가 적절하다고 판단__
   - 차원의 수가 작아질수록 스트레스 값이 증가한다는 원리를 이용
   - __스트레스 값이 가장 크게 감소하는 차원의 수를 찾아 최적의 차원 수 결정__
   - __3차원 이상일 경우 이를 다차원 척도법 결과의 맵으로 표현하기에 한계__ 존재 > 대부분 2차원 또는 3차원 바람직

### 2.2.3.속성-경쟁구조 분석 과정
- 앞서 실시한 단순 경쟁구조 분석을 통해 각 쇼핑몰간의 경쟁구조를 파악할 수 있지만, 어떤 속성 관점에서 경쟁이 이루어지는 알 수 없음


```
import pandas as pd
from sklearn.manifold import MDS
import statsmodels.formula.api as smf
import numpy as np

df = pd.read_csv('/content/gdrive/My Drive/통계/MDS1.csv', sep=',', encoding='CP949')
df1 = df.loc[:,'이미지':'가격만족도']
df1
```

```
# 정규화
df1 = (df1 - df1.min()) / (df1.max() - df1.min())
df1
```

- __자극점과 속성을 같은 평면위에 표현하기 위해__ 정규화 실시
- 정규화
  + 데이터 분포를 조정하는 방법. 데이터 범위 0 ~ 1 사이로 변환
  + (요소값 - 최소값) / (최대값 - 최소값)

```
# 다차원척도법 분석(차원 좌표 값 계산)
clf = MDS(n_components=2, random_state=123).fit(df1)
X_mds = clf.fit_transform(df1)
X_mds
```

```
# 속성 차원 좌표 값 계산
df1['차원1'] = X_mds[:,0]
df1['차원2'] = X_mds[:,1]
model = []
model.append(smf.ols(formula = '이미지 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '접근성 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '서비스 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '친절성 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '편의시설 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '인지도 ~ 차원1+차원2',data=df1).fit())
model.append(smf.ols(formula = '가격만족도 ~ 차원1+차원2',data=df1).fit())

속성 = []
for i in range(0,7):
  속성.append([model[i].params[1],model[i].params[2]])
속성 = np.array(속성)
자극점및속성 = np.hstack([X_mds,속성])
자극점및속성
```

- 속성을 종속변수로, 자극점의 차원 좌표값을 독립변수로 하는 회귀분석 수행
- 독립변수의 계수(벡터의 방향)를 다시 좌표 값으로 하여 속성으로 지정
- 자극점및속성: 자극점의 차원1 좌표, 자극점의 차원2 좌표, 속성의 차원1 좌표, 속성의 차원2 좌표

```
# 속성-경쟁구조 시각화

labels = df.shop
for label,x_count,y_count in zip(labels,자극점및속성[:,0],자극점및속성[:,1]):
  plt.annotate(label, xycoords='data', textcoords='offset points', xy=(x_count,y_count), xytext=(5,-5))

labels2 = df1.columns
for label,x_count, y_count in zip(labels2, 자극점및속성[:,2],자극점및속성[:,3]):
  plt.annotate(label, xycoords='data', textcoords='offset points', xy=(x_count,y_count), xytext=(5,-5))

plt.xlabel('Dim1')
plt.ylabel('Dim2')
plt.scatter(자극점및속성[:,0], 자극점및속성[:,1])   # 자극점 좌표
plt.scatter(자극점및속성[:,2], 자극점및속성[:,3])   # 속성좌표
```

- A쇼핑의 경우 친절성에 상당히 경쟁력있지만, 이미지나 접근성 관점에서는 경쟁력이 약하다는 것을 알 수 있음
